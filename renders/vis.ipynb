```python
RGB_to_mcat40 = {
  (0, 0, 255): 'wall',
  (0, 255, 0): 'floor',
  (0, 255, 255): 'chair',
  (255, 0, 255): 'door',
  (255, 128, 0): 'table',
  (0, 255, 128): 'picture',
  (128, 0, 255): 'cabinet',
  (128, 255, 0): 'cushion',
  (0, 128, 255): 'window',
  (255, 0, 128): 'sofa',
  (128, 0, 0): 'bed',
  (0, 128, 0): 'curtain',
  (0, 0, 128): 'chest_of_drawers',
  (128, 128, 0): 'plant',
  (0, 128, 128): 'sink',
  (128, 0, 128): 'stairs',
  (178, 0, 0): 'ceiling',
  (0, 178, 0): 'toilet',
  (0, 0, 178): 'stool',
  (178, 178, 0): 'towel',
  (0, 178, 178): 'mirror',
  (178, 0, 178): 'tv_monitor',
  (178, 76, 0): 'shower',
  (0, 178, 76): 'column',
  (76, 0, 178): 'bathtub',
  (76, 178, 0): 'counter',
  (0, 76, 178): 'fireplace',
  (178, 0, 76): 'lighting',
  (76, 0, 0): 'beam',
  (0, 76, 0): 'railing',
  (0, 0, 76): 'shelving',
  (76, 76, 0): 'blinds',
  (0, 76, 76): 'gym_equipment',
  (76, 0, 76): 'seating',
  (255, 76, 76): 'board_panel',
  (76, 255, 76): 'furniture',
  (76, 76, 255): 'appliances',
  (255, 255, 76): 'clothes',
  (76, 255, 255): 'objects',
  (255, 76, 255): 'misc',
}
```

```python
data = cv2.imread('output_2n8kARJN3HM_level_0.0.png')
data = cv2.cvtColor(data, cv2.COLOR_BGR2RGB)

class Formatter(object):
    def __init__(self, im):
        self.im = im
    def __call__(self, x, y):
        pix = self.im.get_array()[int(y), int(x)]
        rgb = (float(pix[0]), float(pix[1]), float(pix[2]))
        if rgb in RGB_to_mcat40:
            return RGB_to_mcat40[rgb]
        else:
            return 'invalid'

fig, ax = plt.subplots()
im = ax.imshow(data, interpolation='none')
ax.format_coord = Formatter(im)
plt.show()
```
